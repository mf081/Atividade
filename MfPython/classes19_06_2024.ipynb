```python
# Definindo a classe Animal
class Animal():

    # Construtor da classe
    def __init__(self, name, weight, color):
        self.nome = name
        self.peso = weight
        self.corDoPelo = color

    # Método para imprimir os dados do animal
    def imprimirDados(self):
        return "Nome: %s\nPeso: %.1f\nCor do pelo: %s" % (self.nome, self.peso, self.corDoPelo)

    # Método para fazer o animal emitir um som
    def latir(self, barulho):
        print(barulho)
```

```python

# Instanciando objetos da classe Animal
caozinho = Animal("Marlene", 7.9, "Caramelo")
gato = Animal("Cibele", 4.5, "Cinza")

# Chamando os métodos dos objetos criados
print(caozinho.imprimirDados())
caozinho.latir("Au, Au!")

print(gato.imprimirDados())
gato.latir("Miau!")
```

```python
#criando uma classe chamada Livro
class Livro():

    #Este método vai iniciarlizar cada objeto criado a partir desta classe
    #O nome deste método é __init__
    # (self) é uma referência a cada atributo de um objeto criado a partir desta
    def __init__(self,titulo,isbn):

        # Atributos de cada objeto criado a partir desta classe.
        #o self indica que estes são atributos dos osbjetos
        self.titulo = 'O Monge e o Executivo'
        self.isbn = 9988888
        print("Construtor chamado para criar um objeto desta classe")

    #Métodos são funções, que recebem como parametro atributos do objeto criado
    def imprime(self,titulo,isbn):
        print("Foi criado o livro %s e ISBN %d" %(self.titulo, self.isbn))
```

```python
#Criando uma instancia da classe Livro
Livro1 = Livro()
```
