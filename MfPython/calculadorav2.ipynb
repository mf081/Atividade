```python
def soma(n1, n2):
    return n1 + n2

def subtrai(n1, n2):
    return n1 - n2

def multiplica(n1, n2):
    return n1 * n2

def divide(n1, n2):
    return n1 / n2

def addPercen(n1, n2):
    return n1 + (n1 * (n2 / 100))

def remPercen(n1, n2):
    return n1 - (n1 * (n2 / 100))

def exponenciacao(n1, n2):
    return n1 ** n2

def raiz_quadrada(n1):
    return n1 ** 0.5

while True:
    print("+==============================+")
    print("|     ╔═══════════════════╗    |")
    print("|     ║ MENU DE OPERAÇÕES ║    |")
    print("|     ╚═══════════════════╝    |")
    print("|------------------------------|")
    print("| 1 - Somar (+)                |")
    print("| 2 - Subtrair (-)             |")
    print("| 3 - Multiplicar              |")
    print("| 4 - Dividir                  |")
    print("| 5 - Adicionar Porcentagem    |")
    print("| 6 - Remover Porcentagem      |")
    print("| 7 - Exponenciação            |")
    print("| 8 - Raiz Quadrada            |")
    print("| 0 - Sair                     |")
    print("+==============================+")

    op = int(input())
    if op == 0:
        break
    elif op in [5, 6]:
        n1 = float(input("Primeiro número: "))
        n2 = float(input("Porcentagem: (exemplo: 10) "))
    elif op == 8:
        n1 = float(input("Número: "))
    else:
        n1 = float(input("Primeiro número: "))
        n2 = float(input("Segundo número: "))

    if op == 1:
        print('O resultado da soma é:', soma(n1, n2))
    elif op == 2:
        print('O resultado da subtração é:', subtrai(n1, n2))
    elif op == 3:
        print('O resultado da multiplicação é:', multiplica(n1, n2))
    elif op == 4:
        print('O resultado da divisão é:', divide(n1, n2))
    elif op == 5:
        print('O resultado da adição da porcentagem ao número é:', addPercen(n1, n2))
    elif op == 6:
        print('O resultado da remoção da porcentagem do número é:', remPercen(n1, n2))
    elif op == 7:
        print('O resultado da exponenciação é:', exponenciacao(n1, n2))
    elif op == 8:
        print('O resultado da raiz quadrada é:', raiz_quadrada(n1))
    else:
        print('Por favor, digite uma das opções.')
```
